# Dictionaries

I am going to define a dictionary that creates a relationship between set A and B

```
A = ['Lan', 'Man', 'Cina', 'Lana', 'Rada', 'Vada', 'Wana']
B = ['Kina', 'Wana', 'Pada', 'Cina', 'Rada', 'Jara']

dictionary = dict({a:b for a,b in zip(A, B)})
```

```
actualDictionary = dict({'Vada':'Kina','Rada':'Cina', 'C3P0':'Rada'})

#the only one that really makes sence is Kina and Vada
```

```
newlist2 = [actualDictionary.get(x,x) for x in A]
```

newlist2

This did make some changes to the list! This dictonary is a function, I only change a few things and nothing into the same item (I don't think dictonaries can do that). It is not onto so not bi-jective.

The relation is not an endo relation but it is not symmetric darth vader goes to anakin but not back. Not reflexive since I don't transfer Man back to Man. It is transitive because there is no counter example! Basically I never mapped anything from the range in the domain so it is transitive.
